# Reproducing Fig. 2

```python
from enum import unique
import importlib
import xgi
import matplotlib.pyplot as plt
import random
import numpy as np
import itertools
from HONG import *
import HONG
importlib.reload(HONG)
import csv
from tqdm import tqdm
import csv
from numpy import genfromtxt
import matplotlib as mpl

mpl.style.use('report_style.mplstyle')
```

Parameters of the experiment

```python
prop_committed_B = 0.03
prop_committed_C = 0.1
beta_non_committed = 0.2759
beta_committed = 0.2759
ensemble_size = 10
run_length = 10**5
social_structures = ['InVS15', 'LyonSchool', 'SFHH', 'Thiers13']
```

# Plotting

Plot results for ensemble experiment on linear time axis

```python
social_structure = 'LyonSchool'

fname = f'{social_structure}_B={prop_committed_B}_C={prop_committed_C}_{beta_non_committed}_{beta_committed}_{run_length}_{ensemble_size}'

data = genfromtxt(f'outputs/{fname}.csv', delimiter=',')

print('loaded data')

A_data = data[0::4,:]
B_data = data[1::4,:]
C_data = data[2::4,:]
Mixed_data = data[3::4,:]


A_data_av = np.median(A_data, axis=0)
A_data_25 = np.percentile(A_data, 25, axis=0)
A_data_75 = np.percentile(A_data, 75, axis=0)
B_data_av = np.median(B_data, axis=0)
B_data_25 = np.percentile(B_data, 25, axis=0)
B_data_75 = np.percentile(B_data, 75, axis=0)
C_data_av = np.median(C_data, axis=0)
C_data_25 = np.percentile(C_data, 25, axis=0)
C_data_75 = np.percentile(C_data, 75, axis=0)
Mixed_data_av = np.median(Mixed_data, axis=0)
Mixed_data_25 = np.percentile(Mixed_data, 25, axis=0)
Mixed_data_75 = np.percentile(Mixed_data, 75, axis=0)

t = np.linspace(0, run_length, num=run_length+1, endpoint=True)

plt.figure(1)
plt.plot(t, A_data_av, color='tab:blue', label='A')
plt.fill_between(t, y1=A_data_25, y2=A_data_75, color='tab:blue', alpha=0.2)
plt.plot(t, B_data_av, color='tab:orange', label='B')
plt.fill_between(t, y1=B_data_25, y2=B_data_75, color='tab:orange', alpha=0.2)
plt.plot(t, C_data_av, color='tab:red', label='C')
plt.fill_between(t, y1=C_data_25, y2=C_data_75, color='tab:red', alpha=0.2)
plt.plot(t, Mixed_data_av, color='tab:green', label='Mixed')
plt.fill_between(t, y1=Mixed_data_25, y2=Mixed_data_75, color='tab:green', alpha=0.2)
plt.legend(title=social_structure)
plt.xlabel(r'$t$ / number of interactions')
plt.ylabel(r'$N_{x}(t)$')
plt.legend(title=r'$x$')
plt.savefig(f'figures/{fname}_lintime.pdf')
plt.show()
```

Produces multiple logarithmic plots of individual ensemble experiments

```python
social_structures = {'InVS15':0, 'LyonSchool':0, 'SFHH':0, 'Thiers13':0}

for social_structure in list(social_structures.keys()):
    fname = f'{social_structure}_B={prop_committed_B}_C={prop_committed_C}_{beta_non_committed}_{beta_committed}_{run_length}_{ensemble_size}'

    data = genfromtxt(f'outputs/{fname}.csv', delimiter=',')

    print('loaded data')

    A_data = data[0::4,:]
    B_data = data[1::4,:]
    C_data = data[2::4,:]
    Mixed_data = data[3::4,:]

    N = A_data[0,0]+B_data[0,0]+C_data[0,0]+Mixed_data[0,0]

    A_data = A_data/N
    B_data = B_data/N
    C_data = C_data/N
    Mixed_data = Mixed_data/N

    A_data_av = np.median(A_data, axis=0)
    A_data_25 = np.percentile(A_data, 25, axis=0)
    A_data_75 = np.percentile(A_data, 75, axis=0)
    B_data_av = np.median(B_data, axis=0)
    B_data_25 = np.percentile(B_data, 25, axis=0)
    B_data_75 = np.percentile(B_data, 75, axis=0)
    C_data_av = np.median(C_data, axis=0)
    C_data_25 = np.percentile(C_data, 25, axis=0)
    C_data_75 = np.percentile(C_data, 75, axis=0)
    Mixed_data_av = np.median(Mixed_data, axis=0)
    Mixed_data_25 = np.percentile(Mixed_data, 25, axis=0)
    Mixed_data_75 = np.percentile(Mixed_data, 75, axis=0)

    t = np.linspace(0, run_length, num=run_length+1, endpoint=True)

    plt.figure(1)
    plt.plot(t, A_data_av, color='tab:blue', label='A')
    plt.fill_between(t, y1=A_data_25, y2=A_data_75, color='tab:blue', alpha=0.2)
    plt.plot(t, B_data_av, color='tab:orange', label='B')
    plt.fill_between(t, y1=B_data_25, y2=B_data_75, color='tab:orange', alpha=0.2)
    plt.plot(t, C_data_av, color='tab:red', label='C')
    plt.fill_between(t, y1=C_data_25, y2=C_data_75, color='tab:red', alpha=0.2)
    plt.plot(t, Mixed_data_av, color='tab:green', label='Mixed')
    plt.fill_between(t, y1=Mixed_data_25, y2=Mixed_data_75, color='tab:green', alpha=0.2)
    plt.xscale('log')
    plt.title(r'$\beta$' + f' = {beta_non_committed}')
    plt.xlabel('$t$ / number of interactions')
    plt.ylabel(r'$N_{x}(t)$')
    plt.legend(title=social_structure)
    plt.savefig(f'figures/{fname}_logtime.pdf')
    plt.show()
```

Full reproduction of plots

```python
social_structures = {'InVS15':0, 'LyonSchool':0, 'SFHH':0, 'Thiers13':0}

for social_structure in social_structures:
    fname = f'{social_structure}_B={prop_committed_B}_C={prop_committed_C}_{beta_non_committed}_{beta_committed}_{run_length}_{ensemble_size}'

    data = genfromtxt(f'outputs/{fname}.csv', delimiter=',')

    print('loaded data')
    A_data = data[0::4,:]
    B_data = data[1::4,:]
    C_data = data[2::4,:]
    Mixed_data = data[3::4,:]

    N = A_data[0,0]+B_data[0,0]+C_data[0,0]+Mixed_data[0,0]


    B_data = B_data/N
    
    B_data_av = np.median(B_data, axis=0)
    B_data_25 = np.percentile(B_data, 25, axis=0)
    B_data_75 = np.percentile(B_data, 75, axis=0)
    t = np.linspace(0, run_length, num=run_length+1, endpoint=True)

    plt.plot(t, B_data_av, label = social_structure)
    plt.fill_between(t, y1=B_data_25, y2=B_data_75, alpha=0.2)

    plt.xlabel('$t$ / number of interactions', fontsize=16)
    plt.ylabel(r'$n_{B+B_c}(t)$', fontsize=16)

fname = f'allstructures_{prop_committed}_{beta_non_committed}_{beta_committed}_{run_length}_{ensemble_size}'
plt.legend()
plt.title(r'$\beta$' + f' = {beta_non_committed}')
plt.xscale('log')
plt.xlim((1e2, 1e6))
plt.savefig(f'figures/{fname}_B_logtime.pdf')
plt.show()


for social_structure in social_structures:
    fname = f'{social_structure}_B={prop_committed_B}_C={prop_committed_C}_{beta_non_committed}_{beta_committed}_{run_length}_{ensemble_size}'

    data = genfromtxt(f'outputs/{fname}.csv', delimiter=',')

    print('loaded data')
    A_data = data[0::4,:]
    B_data = data[1::4,:]
    C_data = data[2::4,:]
    Mixed_data = data[3::4,:]

    N = A_data[0,0]+B_data[0,0]+C_data[0,0]+Mixed_data[0,0]


    A_data = A_data/N
    
    A_data_av = np.median(A_data, axis=0)
    A_data_25 = np.percentile(A_data, 25, axis=0)
    A_data_75 = np.percentile(A_data, 75, axis=0)
    t = np.linspace(0, run_length, num=run_length+1, endpoint=True)

    plt.plot(t, A_data_av, label = social_structure)
    plt.fill_between(t, y1=A_data_25, y2=A_data_75, alpha=0.2)

    plt.xlabel('$t$ / number of interactions', fontsize=16)
    plt.ylabel(r'$n_{A}(t)$', fontsize=16)

fname = f'allstructures_{prop_committed}_{beta_non_committed}_{beta_committed}_{run_length}_{ensemble_size}'
plt.legend()
plt.title(r'$\beta$' + f' = {beta_non_committed}')
plt.xscale('log')
plt.xlim((1e2, 1e6))
plt.savefig(f'figures/{fname}_A_logtime.pdf')
plt.show()

for social_structure in social_structures:
    fname = f'{social_structure}_B={prop_committed_B}_C={prop_committed_C}_{beta_non_committed}_{beta_committed}_{run_length}_{ensemble_size}'

    data = genfromtxt(f'outputs/{fname}.csv', delimiter=',')

    print('loaded data')
    A_data = data[0::4,:]
    B_data = data[1::4,:]
    C_data = data[2::4,:]
    Mixed_data = data[3::4,:]

    N = A_data[0,0]+B_data[0,0]+C_data[0,0]+Mixed_data[0,0]


    C_data = C_data/N
    
    C_data_av = np.median(C_data, axis=0)
    C_data_25 = np.percentile(C_data, 25, axis=0)
    C_data_75 = np.percentile(C_data, 75, axis=0)
    t = np.linspace(0, run_length, num=run_length+1, endpoint=True)

    plt.plot(t, C_data_av, label = social_structure)
    plt.fill_between(t, y1=C_data_25, y2=C_data_75, alpha=0.2)

    plt.xlabel('$t$ / number of interactions', fontsize=16)
    plt.ylabel(r'$n_{C+C_c}(t)$', fontsize=16)

fname = f'allstructures_{prop_committed}_{beta_non_committed}_{beta_committed}_{run_length}_{ensemble_size}'
plt.legend()
plt.title(r'$\beta$' + f' = {beta_non_committed}')
plt.xscale('log')
plt.xlim((1e2, 1e6))
plt.savefig(f'figures/{fname}_C_logtime.pdf')
plt.show()



social_structures = {'InVS15':217, 'LyonSchool':218, 'SFHH':255, 'Thiers13':327}


for social_structure in social_structures:
    fname = f'{social_structure}_B={prop_committed_B}_C={prop_committed_C}_{beta_non_committed}_{beta_committed}_{run_length}_{ensemble_size}'

    N = social_structures[social_structure]
    data = genfromtxt(f'outputs/{fname}.csv', delimiter=',')/N

    print('loaded data')
    A_data = data[0::4,:]
    B_data = data[1::4,:]
    C_data = data[2::4,:]
    Mixed_data = data[3::4,:]

    N = A_data[0,0]+B_data[0,0]+C_data[0,0]+Mixed_data[0,0]


    Mixed_data = Mixed_data/N
    
    Mixed_data_av = np.median(Mixed_data, axis=0)
    Mixed_data_25 = np.percentile(Mixed_data, 25, axis=0)
    Mixed_data_75 = np.percentile(Mixed_data, 75, axis=0)
    t = np.linspace(0, run_length, num=run_length+1, endpoint=True)

    plt.plot(t, Mixed_data_av, label = social_structure)
    plt.fill_between(t, y1=Mixed_data_25, y2=Mixed_data_75, alpha=0.2)

    plt.xlabel('$t$ / number of interactions', fontsize=16)
    plt.ylabel(r'$n_{Mixed}(t)$', fontsize=16)

fname = f'allstructures_{prop_committed}_{beta_non_committed}_{beta_committed}_{run_length}_{ensemble_size}'
plt.legend()
plt.xscale('log')
plt.xlim((1e2, 1e6))
plt.title(r'$\beta$' + f' = {beta_non_committed}')
plt.savefig(f'figures/{fname}_Mixed_logtime.pdf')
plt.show()
```
